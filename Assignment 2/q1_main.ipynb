```python
import pandas as pd

df = pd.read_csv('Main.csv')
df
```

```python
#！ To be extended
beta_list = []
for i in range(0,118):
    list1 = df['X'][i:i+24].values.tolist()
    list2 = df['Y'][i:i+24].values.tolist()
    Sxy = Sxx = 0
    for a in range(0,24):
        Sxy += list2[a]*(list1[a]-sum(list1)/len(list1))
        Sxx += (list1[a]-sum(list1)/len(list1))*(list1[a]-sum(list1)/len(list1))
    beta = Sxy/Sxx
    beta_list.append(beta)
beta_list
```
