Then, we will introduce the slack variable in this model. This will include the following steps:
- convert $y$ to binary class, i.e., class 0, class 1, and class 2.
- compute the value of slack variables.

```python
# change y to the binary class
lst_0 = list(y_train==0)
lst_1 = list(y_train==1)
lst_2 = list(y_train==2)
y_train_bi = pd.DataFrame(data=[lst_0, lst_1, lst_2]).T
y_train_bi.columns=['class_0', 'class_1', 'class_2']
y_train_bi.replace(False, -1, inplace=True)
y_train_bi = y_train_bi.astype('int64')
y_train_bi.head()
```

```python
# compute the value of slack variables
df_slack = 1 - y_train_bi*model2.decision_function(X_train)

# then, we drop some unrelated slack variables
def change_unrelated(x):
    if x<0:
        return np.nan
    else:
        return x

df_slack['class_0'] = df_slack['class_0'].apply(lambda x: change_unrelated(x))
df_slack['class_1'] = df_slack['class_1'].apply(lambda x: change_unrelated(x))
df_slack['class_2'] = df_slack['class_2'].apply(lambda x: change_unrelated(x))

df_slack.dropna(how='all', axis=0, inplace=True)
df_slack = df_slack.T

# set index
df_slack['C'] = 0.1
df_slack.reset_index(inplace=True)
df_slack
```

Since now we have completed the whole process of dealing with the slack variables. Now, we will iterate our program to test different C.

```python
for t in range(2, 11):
    # train the model
    c = 0.1*t
    baseline = SVC(C=c, kernel='linear')
    model3 = OneVsRestClassifier(baseline)
    model3.fit(X_train, y_train)

    # coefficient report
    c0_coef = list(model3.coef_[0])
    c1_coef = list(model3.coef_[1])
    c2_coef = list(model3.coef_[2])

    df_coef_2 = pd.DataFrame(columns=['features', 'class_0', 'class_1', 'class_2'])
    df_coef_2['features'] = wines.feature_names
    df_coef_2['class_0'] = c0_coef
    df_coef_2['class_1'] = c1_coef
    df_coef_2['class_2'] = c2_coef

    df_coef_2.loc[len(df_coef_2)]=['intercept', model3.intercept_[0][0], model3.intercept_[1][0], model3.intercept_[2][0]]
    df_coef_2.set_index('features', inplace=True)
    df_coef_2 = df_coef_2.T
    df_coef_2['C'] = c
    df_coef_2.reset_index(inplace=True)
    # merge the coefficient report dataframe
    df_coef = pd.merge(df_coef, df_coef_2, how='outer')

    # support vector indices report
    support_subdict_2 = {}
    support_0 = list(model3.estimators_[0].support_)
    support_1 = list(model3.estimators_[1].support_)
    support_2 = list(model3.estimators_[2].support_)

    support_subdict_2['class_0'] = support_0
    support_subdict_2['class_1'] = support_1
    support_subdict_2['class_2'] = support_2

    support_dict[str(c)] = support_subdict_2
    # report the errors
    train_error = 1-model3.score(X_train, y_train)
    test_error = 1-model3.score(X_test, y_test)

    df_error_2 = pd.DataFrame(columns=['train_error', 'test_error'])
    df_error_2.loc[0, 'train_error'] = train_error
    df_error_2.loc[0, 'test_error'] = test_error
    df_error_2['C'] = c
    # merge the error dataframe
    df_error = pd.merge(df_error, df_error_2, how='outer')
    # convert to binary class
    lst_0 = list(y_train==0)
    lst_1 = list(y_train==1)
    lst_2 = list(y_train==2)
    y_train_bi = pd.DataFrame(data=[lst_0, lst_1, lst_2]).T
    y_train_bi.columns=['class_0', 'class_1', 'class_2']
    y_train_bi.replace(False, -1, inplace=True)
    y_train_bi = y_train_bi.astype('int64')

    # compute slack values
    df_slack_2 = 1 - y_train_bi*model3.decision_function(X_train)
    df_slack_2['class_0'] = df_slack_2['class_0'].apply(lambda x: change_unrelated(x))
    df_slack_2['class_1'] = df_slack_2['class_1'].apply(lambda x: change_unrelated(x))
    df_slack_2['class_2'] = df_slack_2['class_2'].apply(lambda x: change_unrelated(x))

    df_slack_2.dropna(how='all', axis=0, inplace=True)
    df_slack_2 = df_slack_2.T
    df_slack_2['C'] = c
    df_slack_2.reset_index(inplace=True)
    # merge the slack variable dataframe
    df_slack = pd.merge(df_slack, df_slack_2, how='outer')
```
