# 通过cross validation 调整参数α

```python
import pandas as pd
merged_df1 = pd.read_csv('/Users/ryan/Documents/GitHub/CUHKSZ-Course-Projects/FIN 3210/data/Q2_merged.csv')
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler

# 假设merged_df1是你的数据框
# 选择要用作X和y的列
x_columns = ['rmxrf', 'SMB', 'HML', 'Skewness20', 'SharpeRatio20', 'S_VAL_PS', 'NET_ASSETS_TODAY', 'VOL20', 'WVAD']
y_column = 'ret_rf'

# 提取X和y
X = merged_df1[x_columns]
y = merged_df1[y_column]

# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 创建Lasso回归模型
lasso_model = Lasso()

# 设置alpha的候选值
alphas = np.logspace(-8, 4, 9)

# 设置交叉验证折数
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# 设置参数网格
param_grid = {'alpha': alphas}

# 创建GridSearchCV对象
grid_search = GridSearchCV(lasso_model, param_grid, cv=cv, scoring='neg_mean_squared_error', n_jobs=-1)

# 在整个数据集上拟合GridSearchCV对象
grid_search.fit(X_scaled, y)

# 输出最佳参数值
best_alpha = grid_search.best_params_['alpha']
print("best_alpha:", best_alpha)

# 输出最佳模型
best_model = grid_search.best_estimator_
print("best_model:", best_model)

# 输出最佳模型的系数
best_coefficients = pd.Series(best_model.coef_, index=x_columns)
print("best_coefficients:")
print(best_coefficients)
```

```python
## 对比参数α的为0.01和0.0001时的模型表现

from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 假设 merged_df1 是你的数据框
# 选择要用作 X 和 y 的列
x_columns = ['rmxrf', 'SMB', 'HML', 'Skewness20', 'SharpeRatio20', 'S_VAL_PS', 'NET_ASSETS_TODAY', 'VOL20', 'WVAD']
y_column = 'ret_rf'

# 提取 X 和 y
X = merged_df1[x_columns]
y = merged_df1[y_column]

# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 划分数据集为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 不同的 alpha 值
alpha_values = [0.01, 0.001,0.0001]

for alpha in alpha_values:
    # 创建 Lasso 回归模型
    lasso_model = Lasso(alpha=alpha)

    # 拟合模型
    lasso_model.fit(X_train, y_train)

    # 输出模型系数
    coefficients = pd.Series(lasso_model.coef_, index=x_columns)
    print(f"\n在 alpha={alpha} 时的系数:")
    print(coefficients)

    # 在测试集上进行预测
    y_pred = lasso_model.predict(X_test)

    # 计算均方误差 (MSE)
    mse = mean_squared_error(y_test, y_pred)
    print(f"在 alpha={alpha} 时的均方误差 (MSE): {mse}")
```

```python
# 选择指定 stk 的数据
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

selected_stks = merged_df1['stk'].unique().tolist()
for stock_id in selected_stks:
    stk_data = merged_df1[merged_df1['stk'] ==  stock_id]

    # 提取 X 和 y
    X_stk = stk_data[x_columns]
    y_stk = stk_data[y_column]
    dates_stk = stk_data['date']

    # 数据标准化
    X_stk_scaled = scaler.transform(X_stk)
    
    # 在测试集上进行预测
    y_pred_stk = lasso_model.predict(X_stk_scaled)
    
    mse = mean_squared_error(y_stk, y_pred_stk)
    print(mse)
    # 绘制实际值 vs 预测值的对比图
    data_stk_datetime = pd.to_datetime(dates_stk)

    plt.figure(figsize=(16, 4))   
    plt.plot(data_stk_datetime, y_stk, label='Actual')
    plt.plot(data_stk_datetime, y_pred_stk, label='Predicted', color='orange')

    plt.title(f"{stock_id} - Actual vs Predicted Return")
    plt.xlabel('Date')
    plt.ylabel('Ret-rf')
    plt.legend()

    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.gca().xaxis.set_major_locator(mdates.YearLocator())

    plt.xticks(rotation=45)

    plt.show()
```
